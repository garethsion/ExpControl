```python
import ExpControl
import numpy as np
import pandas as pd
import pickle
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
plt.rcParams['lines.linewidth'] = 3.0
plt.rcParams['font.size'] = 22.0
plt.rcParams['xtick.labelsize'] = 22
plt.rcParams['ytick.labelsize'] = 22
plt.rcParams['axes.labelsize'] = 26
plt.style.use('seaborn-white')
```

```python
meas = ExpControl.Measurements(instr=['vna'])
```

```python
meas.setup_vna(power=-30,avgs = 1,measure='S21')
```

```python
trace_wb = meas.trace(0e09,20.0e09,npts=20001,bandwidth=1000)
```

```python
datadir = 'data/2019_11_21_Nb_50nm_on_Si_250um_w_Bi_10^18'

df_wb = pd.DataFrame(data={'freq':trace_wb[0],'ch1':trace_wb[1][0], 'ch2':trace_wb[1][1]})
df_wb.to_pickle(datadir+'/S21_0G_20G.pkl')
```

```python
trace_8p9G_9p1G = meas.trace(8.9e09,9.1e09,npts=20001,bandwidth=1000)
```

```python
df = pd.DataFrame(data={'freq':trace_8p9G_9p1G[0],'ch1':trace_8p9G_9p1G[1][0], 'ch2':trace_8p9G_9p1G[1][1]})
df.to_pickle(datadir+'/S21_8p9G_9p1G.pkl')
```

```python
trace_9p0G_9p01825G = meas.trace(9.0e09,9.01825e09,npts=20001,bandwidth=1000)
```

```python
df = pd.DataFrame(data={'freq':trace_9p0G_9p01825G[0],'ch1':trace_9p0G_9p01825G[1][0], 'ch2':trace_9p0G_9p01825G[1][1]})
df.to_pickle(datadir+'/S21_9p0G_9p01825G.pkl')
```
